```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression #Logistic Regression helps to classify
from sklearn.metrics import mean_squared_error, r2_score #squared errors help to magnify the large errors r2 score helps to give difference between predicted and actual value
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("/content/income_data.csv")
x=df[['income']]
y=df['happiness']
plt.scatter(x,y,color='blue')
plt.title("Income vs. Happiness")
plt.xlabel("Income")
plt.ylabel("Happiness")
plt.show()

#Train Test Split
X_train, X_test, y_train, y_test=train_test_split(x, y, test_size=0.2, random_state=42) # helps in reproductivity to create similar output
print("train set size:", len(X_train))
print("test set size:", len(X_test))

#Defining linear regression model
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
#visualizing with a regression model

predicted_happiness=model.predict(x)
plt.scatter(x,y,color='red', label='Actual Happiness')
plt.plot(x,predicted_happiness, color='green', label='Regression Line')
plt.scatter(x,predicted_happiness,color='blue', label='Predicted Happiness')
plt.xlabel('Income')
plt.ylabel('Happiness')
plt.title('Income vs. Happiness')
plt.legend()
plt.show()
```

```python
#evaluate model performance

y_train_pred=model.predict(X_train)
y_test_pred= model.predict(X_test)

#MSE

train_mse=mean_squared_error(y_train, y_train_pred)
test_mse=mean_squared_error(y_test, y_test_pred)

#R Squared

train_r2=r2_score(y_train, y_train_pred)
test_r2=r2_score(y_test, y_test_pred)
print("Train MSE:", train_mse)
print("Test MSE:", test_mse)
print("Train R-Squared:", train_r2)
print("Test R-squared:", test_r2)

#Predicting (Inferencing)
print(model.predict([[5]]))
```
