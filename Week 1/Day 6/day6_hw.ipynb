```python
import io
csv_data="""
Product,Price
Apple,1.0
Banana,0.5
Orange,0.75
"""
df=pd.read_csv(io.StringIO(csv_data))
print(df)
sum_price=df["Price"].sum()
print(sum_price)
```

```python
import matplotlib.pyplot as plt
x=[0, 1, 2, 3]
y=[0, 1, 4, 9]
plt.plot(x,y,marker='o')
plt.title("Square of Numbers")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True)
plt.savefig("squareplot.png")
plt.show()
```

```python
data = {
    "City": ["London", "Paris", "Tokyo"],
    "Population": [9.0, 2.1,37.4]
}
df = pd.DataFrame(data)

plt.bar(df["City"],df["Population"])
plt.title("Bar Plot")
plt.xlabel("City")
plt.ylabel("Population")
plt.show()
plt.savefig("Populationplot.png")
```
